 # Phân tích Án lệ: Thống kê vi phạm theo Điều và Chương

 Notebook này xử lý danh sách các file JSON (định dạng `stage2.json`)

 để trích xuất các "Điều" từ trường `QUYET_DINH`. Sau đó ánh xạ

 các "Điều" này sang "Chương" tương ứng thông qua `law_doc.json`

 và hiển thị những Điều, Chương bị vi phạm nhiều nhất (mỗi case chỉ đếm 1 lần cho 1 Chương).

```python
import json
import re
import glob
import os
from collections import Counter
```

 ## 1. Load dữ liệu ánh xạ Luật

 Đọc `law_doc.json` để tạo từ điển ánh xạ từ "Điều" sang "Chương".

```python
def build_dieu_to_chuong_map(law_doc_path):
    dieu_to_chuong = {}
    try:
        with open(law_doc_path, 'r', encoding='utf-8') as f:
            law_data = json.load(f)
            
        for chuong in law_data:
            if chuong.get("type") == "Chương":
                chuong_idx = str(chuong.get("index", "Unknown"))
                # Lặp qua các Điều trong Chương
                for dieu in chuong.get("Điều", []):
                    if dieu.get("type") == "Điều":
                        dieu_idx = str(dieu.get("index", ""))
                        if dieu_idx:
                            dieu_to_chuong[dieu_idx] = chuong_idx
    except Exception as e:
        print(f"Lỗi khi load {law_doc_path}: {e}")
        
    return dieu_to_chuong

LAW_DOC_PATH = 'law_doc.json'
dieu_to_chuong = build_dieu_to_chuong_map(LAW_DOC_PATH)
print(f"Đã load thành công ánh xạ cho {len(dieu_to_chuong)} Điều.")
```

 ## 2. Trích xuất "Điều" từ QUYET_DINH

```python
def extract_dieu_from_text(text):
    if not text:
        return []
    # Tìm tất cả các số/chữ theo sau chữ "Điều" hoặc "điều"
    matches = re.findall(r'(?i)điều\s+(\d+[a-zA-Z]*)', text)
    return matches
```

 ## 3. Xử lý các file JSON án lệ

 Đếm số lượng Điều và Chương. Mỗi Chương chỉ đếm 1 lần cho mỗi vụ án, cho dù vụ án đó có vi phạm nhiều Điều thuộc cùng Chương đó.

```python
CASELAW_FILES_PATTERN = 'extracted_fields/*.json'  

dieu_counter = Counter()
chuong_counter = Counter()

# Loại trừ file law_doc.json để không bị đọc nhầm như file án lệ
case_files = [f for f in glob.glob(CASELAW_FILES_PATTERN) if not f.endswith('law_doc.json')]

print(f"Tìm thấy {len(case_files)} file án lệ để xử lý.\n")

for file_path in case_files:
    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            case_data = json.load(f)
            
        quyet_dinh_text = case_data.get("QUYET_DINH", "")
        
        # Trích xuất danh sách Điều
        extracted_dieus = extract_dieu_from_text(quyet_dinh_text)
        
        # Dùng set để loại bỏ các Điều trùng lặp trong cùng 1 case
        unique_dieus_in_case = set(extracted_dieus)
        
        # Tạo set để lưu các Chương bị vi phạm trong case này (chống đếm trùng)
        unique_chuongs_in_case = set()
        
        for dieu in unique_dieus_in_case:
            # Đếm số vụ án vi phạm Điều này
            dieu_counter[dieu] += 1
            
            # Tra cứu Chương tương ứng
            chuong = dieu_to_chuong.get(dieu)
            if chuong:
                unique_chuongs_in_case.add(chuong)
            else:
                unique_chuongs_in_case.add("Unknown/Unmapped")
                
        # Sau khi tổng hợp xong tất cả các Chương của case này, tiến hành cộng vào chuong_counter
        for chuong in unique_chuongs_in_case:
            chuong_counter[chuong] += 1
                
    except json.JSONDecodeError:
        pass # Bỏ qua lỗi JSON
    except Exception as e:
        print(f"Lỗi khi xử lý {file_path}: {e}")
```

 ## 4. Hiển thị kết quả

```python
TOP_N = 20

print("="*50)
print(f"🔝 TOP {TOP_N} MOST VIOLATED ARTICLES (ĐIỀU)")
print("="*50)
if not dieu_counter:
    print("Không tìm thấy Điều nào.")
else:
    for rank, (dieu, count) in enumerate(dieu_counter.most_common(TOP_N), 1):
        chuong = dieu_to_chuong.get(dieu, "Unknown")
        print(f"{rank}. Điều {dieu:<5} (Chương {chuong:<3}) : {count} cases")

print("\n" + "="*50)
print(f"🔝 TOP {TOP_N} MOST VIOLATED CHAPTERS (CHƯƠNG)")
print("="*50)
if not chuong_counter:
    print("Không tìm thấy Chương nào.")
else:
    for rank, (chuong, count) in enumerate(chuong_counter.most_common(TOP_N), 1):
        print(f"{rank}. Chương {chuong:<15} : {count} cases")
```

```python
import os
import shutil

# 👉 Change this variable to your desired Chapter
TARGET_CHAPTER = "XXII" 

# Create the directory if it doesn't exist
output_dir = f"Chuong_{TARGET_CHAPTER}"
os.makedirs(output_dir, exist_ok=True)

copied_count = 0

print(f"Bắt đầu lọc và copy các vụ án liên quan đến Chương {TARGET_CHAPTER}...")

for file_path in case_files:
    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            case_data = json.load(f)
            
        quyet_dinh_text = case_data.get("QUYET_DINH", "")
        
        # Lấy danh sách Điều trong phần Quyết định
        extracted_dieus = extract_dieu_from_text(quyet_dinh_text)
        
        # Kiểm tra xem có Điều nào thuộc Chương mục tiêu không
        has_target_chuong = False
        for dieu in set(extracted_dieus):
            chuong = dieu_to_chuong.get(dieu)
            if chuong == TARGET_CHAPTER:
                has_target_chuong = True
                break  # Chỉ cần 1 Điều thuộc Chương này là đủ điều kiện copy
                
        # Nếu file có chứa Điều thuộc Chương mục tiêu, tiến hành copy
        if has_target_chuong:
            destination_path = os.path.join(output_dir, os.path.basename(file_path))
            shutil.copy2(file_path, destination_path)
            copied_count += 1
            
    except json.JSONDecodeError:
        pass # Bỏ qua lỗi JSON
    except Exception as e:
        print(f"Lỗi khi xử lý {file_path}: {e}")

print("="*50)
print(f"✅ Đã copy thành công {copied_count} file vào thư mục '{output_dir}'.")
print("="*50)
```

```python
def get_unique_names(data_list, name_key):
    """
    Extracts, normalizes, and returns a set of unique names from a list of dicts.
    """
    if not isinstance(data_list, list):
        return set()
    
    names = set()
    for item in data_list:
        if isinstance(item, dict):
            name = item.get(name_key, "")
            if isinstance(name, str) and name.strip():
                # Normalize name: convert to lowercase and remove redundant spaces
                normalized_name = " ".join(name.lower().split())
                names.add(normalized_name)
    return names

# 👉 Define the folder containing the JSON files you want to check.
CHECK_FOLDER = "filled_Chuong_XXII" 

# Find all json files in the target folder
json_files_to_check = glob.glob(os.path.join(CHECK_FOLDER, '*.json'))

total_files = 0
mismatch_files = []

print(f"Bắt đầu kiểm tra số lượng bị cáo (Unique) trong thư mục '{CHECK_FOLDER}'...")

for file_path in json_files_to_check:
    if file_path.endswith('law_doc.json'):
        continue

    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            data = json.load(f)
        
        # 1. Get unique defendants in THONG_TIN_CHUNG -> Thong_Tin_Bi_Cao (Key: "Ho_Ten")
        thong_tin_chung = data.get("THONG_TIN_CHUNG", {})
        thong_tin_bi_cao = thong_tin_chung.get("Thong_Tin_Bi_Cao", [])
        names_thong_tin = get_unique_names(thong_tin_bi_cao, "Ho_Ten")
        
        # 2. Get unique defendants in De_Nghi_Cua_Vien_Kiem_Sat (Key: "Bi_Cao")
        de_nghi = data.get("De_Nghi_Cua_Vien_Kiem_Sat", [])
        names_de_nghi = get_unique_names(de_nghi, "Bi_Cao")
        
        # 3. Get unique defendants in PHAN_QUYET_CUA_TOA_SO_THAM (Key: "Bi_Cao")
        phan_quyet = data.get("PHAN_QUYET_CUA_TOA_SO_THAM", [])
        names_phan_quyet = get_unique_names(phan_quyet, "Bi_Cao")
        
        count_thong_tin = len(names_thong_tin)
        count_de_nghi = len(names_de_nghi)
        count_phan_quyet = len(names_phan_quyet)
        
        # Check if the counts match
        if not (count_thong_tin == count_de_nghi == count_phan_quyet):
            mismatch_files.append({
                "file": os.path.basename(file_path),
                "counts": (count_thong_tin, count_de_nghi, count_phan_quyet),
                "names_diff": {
                    "Thong_Tin": list(names_thong_tin),
                    "De_Nghi": list(names_de_nghi),
                    "Phan_Quyet": list(names_phan_quyet)
                }
            })
        
        total_files += 1

    except json.JSONDecodeError:
        pass # Skip invalid JSON files
    except Exception as e:
        print(f"Lỗi khi xử lý {file_path}: {e}")

# Display Results
print("="*70)
print("📊 KẾT QUẢ KIỂM TRA LỆCH SỐ LƯỢNG BỊ CÁO (UNIQUE DEFENDANT MISMATCH)")
print("="*70)

if total_files == 0:
    print("Không tìm thấy file JSON nào hợp lệ trong thư mục để kiểm tra.")
else:
    mismatch_count = len(mismatch_files)
    mismatch_percentage = (mismatch_count / total_files) * 100
    
    print(f"Tổng số file đã kiểm tra: {total_files}")
    print(f"Số file bị lệch: {mismatch_count}")
    print(f"Tỷ lệ lỗi (Mismatch rate): {mismatch_percentage:.2f}%\n")
    
    if mismatch_count > 0:
        print("🛑 DANH SÁCH CÁC FILE LỖI:")
        for item in mismatch_files:
            counts = item['counts']
            print(f"\n- {item['file']}")
            print(f"  Số lượng: Thong_Tin: {counts[0]}, De_Nghi: {counts[1]}, Phan_Quyet: {counts[2]}")
            # I added a print out of the extracted names here so you can easily see WHY they didn't match
            print(f"  Tên trích xuất:")
            print(f"    + Thong_Tin : {item['names_diff']['Thong_Tin']}")
            print(f"    + De_Nghi   : {item['names_diff']['De_Nghi']}")
            print(f"    + Phan_Quyet: {item['names_diff']['Phan_Quyet']}")
    else:
        print("✅ Tuyệt vời! Tất cả các file đều khớp số lượng bị cáo duy nhất ở cả 3 trường.")
```

```python
empty_name_files = set()
empty_name_details = []

def check_empty_names(data_list, name_key, source_name, file_name):
    """Kiểm tra xem field tên có bị null hoặc chuỗi rỗng không."""
    if not isinstance(data_list, list):
        return
    
    for idx, item in enumerate(data_list):
        if isinstance(item, dict):
            name = item.get(name_key)
            # Kiểm tra: name là None (null trong json) hoặc là chuỗi nhưng rỗng (sau khi strip)
            if name is None or (isinstance(name, str) and not name.strip()):
                empty_name_files.add(file_name)
                empty_name_details.append({
                    "file": file_name,
                    "source": source_name,
                    "index": idx,
                    "key": name_key,
                    "value_found": name
                })

print(f"Bắt đầu kiểm tra tên bị cáo (Null/Empty) trong thư mục '{CHECK_FOLDER}'...")

for file_path in json_files_to_check:
    if file_path.endswith('law_doc.json'):
        continue

    file_name = os.path.basename(file_path)
    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            data = json.load(f)
        
        # 1. Kiểm tra THONG_TIN_CHUNG -> Thong_Tin_Bi_Cao (Key: "Ho_Ten")
        thong_tin_chung = data.get("THONG_TIN_CHUNG", {})
        thong_tin_bi_cao = thong_tin_chung.get("Thong_Tin_Bi_Cao", [])
        check_empty_names(thong_tin_bi_cao, "Ho_Ten", "Thong_Tin_Bi_Cao", file_name)
        
        # 2. Kiểm tra De_Nghi_Cua_Vien_Kiem_Sat (Key: "Bi_Cao")
        de_nghi = data.get("De_Nghi_Cua_Vien_Kiem_Sat", [])
        check_empty_names(de_nghi, "Bi_Cao", "De_Nghi_Cua_Vien_Kiem_Sat", file_name)
        
        # 3. Kiểm tra PHAN_QUYET_CUA_TOA_SO_THAM (Key: "Bi_Cao")
        phan_quyet = data.get("PHAN_QUYET_CUA_TOA_SO_THAM", [])
        check_empty_names(phan_quyet, "Bi_Cao", "PHAN_QUYET_CUA_TOA_SO_THAM", file_name)

    except json.JSONDecodeError:
        pass # Skip invalid JSON files
    except Exception as e:
        print(f"Lỗi khi xử lý {file_path}: {e}")

# Display Results
print("="*70)
print("📊 KẾT QUẢ KIỂM TRA TÊN BỊ CÁO TRỐNG (NULL/EMPTY NAMES)")
print("="*70)

total_files_checked = len([f for f in json_files_to_check if not f.endswith('law_doc.json')])

if total_files_checked == 0:
    print("Không tìm thấy file JSON nào hợp lệ trong thư mục để kiểm tra.")
else:
    empty_error_count = len(empty_name_files)
    empty_error_rate = (empty_error_count / total_files_checked) * 100
    
    print(f"Tổng số file đã kiểm tra: {total_files_checked}")
    print(f"Số file chứa tên bị cáo trống: {empty_error_count}")
    print(f"Tỷ lệ lỗi: {empty_error_rate:.2f}%\n")
    
    if empty_error_count > 0:
        print("🛑 CHI TIẾT CÁC LỖI TÌM THẤY:")
        for detail in empty_name_details:
            print(f"- File: {detail['file']} | Nguồn: {detail['source']} | Vị trí (index): {detail['index']} | Giá trị: {detail['value_found']}")
    else:
        print("✅ Tuyệt vời! Không có bị cáo nào bị thiếu tên (Null/Empty) trong các file.")
```

```python
def load_valid_dieus(law_doc_path):
    valid_dieus = set()
    try:
        with open(law_doc_path, 'r', encoding='utf-8') as f:
            law_data = json.load(f)
            for chuong in law_data:
                if chuong.get("type") == "Chương":
                    for dieu in chuong.get("Điều", []):
                        if dieu.get("type") == "Điều":
                            dieu_idx = str(dieu.get("index", "")).strip()
                            if dieu_idx:
                                valid_dieus.add(dieu_idx)
    except Exception as e:
        print(f"Lỗi khi load {law_doc_path}: {e}")
    return valid_dieus

LAW_DOC_PATH = 'law_doc.json'
valid_dieus = load_valid_dieus(LAW_DOC_PATH)
print(f"Đã load {len(valid_dieus)} Điều hợp lệ từ {LAW_DOC_PATH}.\n")

# 2. Setup checking directory
CHECK_FOLDER = "filled_Chuong_XXII" 
json_files_to_check = glob.glob(os.path.join(CHECK_FOLDER, '*.json'))

invalid_references = []
files_with_errors = set()
total_rules_checked = 0

def check_can_cu_dieu_luat(can_cu_list, source_name, file_name):
    """Helper function to validate the Can_Cu_Dieu_Luat list."""
    global total_rules_checked
    if not isinstance(can_cu_list, list):
        return
        
    for rule in can_cu_list:
        bo_luat = str(rule.get("Bo_Luat_Va_Van_Ban_Khac", "")).strip()
        
        # Case-insensitive check for "Bộ luật Hình sự"
        if "bộ luật hình sự" in bo_luat.lower():
            total_rules_checked += 1
            dieu = str(rule.get("Dieu", "")).strip()
            
            if not dieu:
                invalid_references.append({
                    "file": file_name,
                    "source": source_name,
                    "error": "Thiếu giá trị 'Điều'",
                    "rule": rule
                })
                files_with_errors.add(file_name)
            elif dieu not in valid_dieus:
                invalid_references.append({
                    "file": file_name,
                    "source": source_name,
                    "error": f"Điều '{dieu}' không tồn tại trong law_doc.json",
                    "rule": rule
                })
                files_with_errors.add(file_name)

print(f"Bắt đầu kiểm tra tính hợp lệ của các Điều luật trong thư mục '{CHECK_FOLDER}'...")

for file_path in json_files_to_check:
    if file_path.endswith('law_doc.json'):
        continue

    file_name = os.path.basename(file_path)
    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            data = json.load(f)
        
        # Check De_Nghi_Cua_Vien_Kiem_Sat
        de_nghi_list = data.get("De_Nghi_Cua_Vien_Kiem_Sat", [])
        if isinstance(de_nghi_list, list):
            for item in de_nghi_list:
                check_can_cu_dieu_luat(item.get("Can_Cu_Dieu_Luat", []), "De_Nghi_Cua_Vien_Kiem_Sat", file_name)
        
        # Check PHAN_QUYET_CUA_TOA_SO_THAM
        phan_quyet_list = data.get("PHAN_QUYET_CUA_TOA_SO_THAM", [])
        if isinstance(phan_quyet_list, list):
            for item in phan_quyet_list:
                check_can_cu_dieu_luat(item.get("Can_Cu_Dieu_Luat", []), "PHAN_QUYET_CUA_TOA_SO_THAM", file_name)

    except json.JSONDecodeError:
        pass # Skip invalid JSON files
    except Exception as e:
        print(f"Lỗi khi xử lý {file_path}: {e}")

# 3. Display Results
print("="*70)
print("📊 KẾT QUẢ KIỂM TRA ĐIỀU LUẬT (BỘ LUẬT HÌNH SỰ)")
print("="*70)

if total_rules_checked == 0:
    print("Không tìm thấy trích dẫn nào liên quan đến 'Bộ luật Hình sự' để kiểm tra.")
else:
    error_rate = (len(invalid_references) / total_rules_checked) * 100
    
    print(f"Tổng số file đã duyệt: {len(json_files_to_check)}")
    print(f"Tổng số lần trích dẫn 'Bộ luật Hình sự' được kiểm tra: {total_rules_checked}")
    print(f"Số lượng trích dẫn KHÔNG HỢP LỆ: {len(invalid_references)}")
    print(f"Số file chứa lỗi: {len(files_with_errors)}")
    print(f"Tỷ lệ trích dẫn lỗi: {error_rate:.2f}%\n")

    if invalid_references:
        print("🛑 CHI TIẾT CÁC LỖI TÌM THẤY:")
        for idx, err in enumerate(invalid_references, 1):
            print(f"\n{idx}. File: {err['file']}")
            print(f"   Vị trí: {err['source']}")
            print(f"   Lỗi: {err['error']}")
            print(f"   Dữ liệu gốc: {err['rule']}")
    else:
        print("✅ Tuyệt vời! Tất cả các trích dẫn Bộ luật Hình sự đều tồn tại trong law_doc.json.")
```

```python
import os
import glob
import json
import re

# We condensed the 40+ phrases into 5 robust regex patterns.
# \s+ handles multiple spaces, tabs, or newlines between words.
# (?:...) creates an optional or interchangeable group without capturing it.
# Character classes like [ậaâ] catch common typing/OCR diacritic errors (e.g., kết luận vs kết luân).
REGEX_PATTERNS = [
    # 1. The Core Evaluation: (đã/có) đủ cơ sở/căn cứ/chứng cứ (để) kết luận/xác định/khẳng định
    r"(?:đã\s+)?(?:có\s+)?đủ\s+(?:c[ơo]\s*s[ởo]|c[ăa]n\s*c[ứu]|chứng\s*cứ)\s+(?:để\s+)?(?:kết\s*lu[ậaâ]n|xác\s*định|khẳng\s*định)",
    
    # 2. The Trial Panel: hội đồng xét xử / hđxx (nhận/xét) thấy
    r"(?:hội\s*đồng\s*xét\s*xử|hđxx)\s*(?:nhận|xét)?\s*thấy",
    
    # 3. The "Xét thấy" variations: xét thấy (có) đủ cơ sở/căn cứ
    r"xét\s*thấy\s*(?:có\s+)?đủ\s+(?:c[ơo]\s*s[ởo]|c[ăa]n\s*c[ứu])",
    
    # 4. Elements of crime: (đã/có) đủ yếu tố cấu thành
    r"(?:đã\s+|có\s+)?đủ\s*yếu\s*tố\s*cấu\s*thành",
    
    # 5. Context starters: xét hành vi phạm tội / xét lời khai
    r"xét\s*(?:hành\s*vi\s*phạm\s*tội|lời\s*khai)",
    
]

# Compile patterns once for faster execution across hundreds of files
COMPILED_PATTERNS = [re.compile(p, re.IGNORECASE) for p in REGEX_PATTERNS]

CHECK_FOLDER = "filled_Chuong_XXII" # Thay bằng tên folder gốc của bạn
CHECKED_FOLDER = f"checkd_{CHECK_FOLDER}"

files_to_search = glob.glob(os.path.join(CHECKED_FOLDER, '*.json'))

contains_any_count = 0
contains_none_count = 0

files_containing_any = []
files_containing_none = []

print("Bắt đầu tìm kiếm bằng Regex patterns...")
print(f"Trong thư mục: '{CHECKED_FOLDER}'...\n")

for file_path in files_to_search:
    if file_path.endswith('law_doc.json'):
        continue
        
    file_name = os.path.basename(file_path)
    
    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            data = json.load(f)
            
        nhan_dinh = data.get("NHAN_DINH_CUA_TOA_AN", {})
        
        # Gom tất cả văn bản trong NHAN_DINH_CUA_TOA_AN lại thành 1 chuỗi
        combined_text = ""
        if isinstance(nhan_dinh, dict):
            combined_text = " ".join(str(v) for v in nhan_dinh.values())
        elif isinstance(nhan_dinh, str): 
            combined_text = nhan_dinh
            
        # Kiểm tra xem có text nào khớp với BẤT KỲ pattern nào không
        # Chúng ta không cần .lower() nữa vì đã dùng re.IGNORECASE khi compile
        has_any = any(pattern.search(combined_text) for pattern in COMPILED_PATTERNS)
        
        if has_any:
            contains_any_count += 1
            files_containing_any.append(file_name)
        else:
            contains_none_count += 1
            files_containing_none.append(file_name)
            
    except json.JSONDecodeError:
        print(f"Lỗi đọc JSON: {file_name}")
    except Exception as e:
        print(f"Lỗi khi xử lý {file_name}: {e}")

# Display Results
print("="*75)
print(f"📊 KẾT QUẢ TÌM KIẾM BẰNG REGEX")
print("="*75)

total_searched = len([f for f in files_to_search if not f.endswith('law_doc.json')])

if total_searched == 0:
    print(f"Thư mục '{CHECKED_FOLDER}' trống hoặc không tồn tại file hợp lệ.")
else:
    print(f"Tổng số file đã kiểm tra: {total_searched}")
    print(f"✅ Số file chứa ÍT NHẤT MỘT cụm từ : {contains_any_count} ({(contains_any_count/total_searched)*100:.1f}%)")
    print(f"❌ Số file KHÔNG chứa BẤT KỲ cụm từ nào: {contains_none_count} ({(contains_none_count/total_searched)*100:.1f}%)\n")
    
    if contains_none_count > 0:
        print("🛑 DANH SÁCH FILE KHÔNG KHỚP VỚI BẤT KỲ PATTERN NÀO (NONE):")
        # In ra tối đa 20 file để tránh làm trôi màn hình console
        for idx, f in enumerate(files_containing_none[:20], 1):
            print(f"  {idx}. {f}")
        if contains_none_count > 30:
            print(f"  ... và {contains_none_count - 30} file khác.")
    else:
        print("✅ Tuyệt vời! Tất cả các file đều khớp với ít nhất một pattern.")
```
